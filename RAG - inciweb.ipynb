```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/home/centrox_ai/Desktop/ABDULLAH/llama2/rag/1.pdf")
pages = loader.load_and_split()
```

```python
# Load web page# Split into chunks 
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500, chunk_overlap=100)
all_splits = text_splitter.split_documents(pages)
```

```python
# Split into chunks 
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500, chunk_overlap=100)
all_splits = text_splitter.split_documents(pages)
```

```python
len(all_splits)
```

```python
# Embed and store
from langchain.vectorstores import Chroma
from langchain.embeddings import GPT4AllEmbeddings
# from langchain.embeddings import OllamaEmbeddings # We can also try Ollama embeddings
vectorstore = Chroma.from_documents(documents=all_splits,
                                    embedding=GPT4AllEmbeddings())
```

```python
question = "what is the name of the book?"
docs = vectorstore.similarity_search(question)
len(docs)
docs
```

```python
# RAG prompt 
from langchain import hub
QA_CHAIN_PROMPT = hub.pull("rlm/rag-prompt")
```

```python
# connect with the LLM, in this case llama2
from langchain.llms import Ollama
from langchain.callbacks.manager import CallbackManager
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler                                  
llm = Ollama(model="llama3", 
             callback_manager = CallbackManager([StreamingStdOutCallbackHandler()]))
```

```python
# QA chain
from langchain.chains import RetrievalQA
qa_chain = RetrievalQA.from_chain_type(
    llm,
    retriever=vectorstore.as_retriever(),
    chain_type_kwargs={"prompt": QA_CHAIN_PROMPT},
)
```

```python
question = "tell me about W. S. Ferguson"
result = qa_chain({"query": question})
```
